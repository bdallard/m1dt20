## Data Clean 
Les données nettoyées situées dans le dossier /clean à la racine de ce notebook. Pour plus de details consulter le script **raw_cleaning.py** à la racine. 

```python
#charger le dataset clean
dataset = pd.read_csv("../data/listings_final.csv", sep=";")
```

```python
print(dataset.head)
```

```python
#définition de la variable de travail 
X = np.asarray(dataset.loc[:, ["person_capacity", "beds", "bedrooms", "bathrooms"]].values, dtype="float")
print(X)
```

```python
#calcul de la moyenne 
means = sum(X) / X.shape[0]
#calcul de la variance 
variance = sum((X - means)**2) / X.shape[0]
std = np.sqrt(variance)
```

```python
#calcul de la covariance 
covar = np.zeros((X.shape[1], X.shape[1]))  # Matrice de covariance
#calcul de la corrélation 
corr = np.zeros((X.shape[1], X.shape[1]))  # Matrice de corrélation

#affichage de la variance 
print(variance)
```

## Construction des matrices 

```python
for i in range(X.shape[1]):
	for j in range(i+1):
		covar[i, j] = covar[j, i] = np.dot(X[:,i], X[:,j]) / X.shape[0] - means[i] * means[j]
		corr[i, j] = corr[j, i] = covar[i, j] / (std[i] * std[j])
	

print(means)
print("Matrice des covariances :")
print(covar)
print("Matrice des corrélations :")
print(corr)
```
